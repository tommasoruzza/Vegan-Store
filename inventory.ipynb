```python
#Importing the necessary modules
from product import *

import pickle

class Inventory:
    
    """
    Class to represent an inventory of products.
    """
    def __init__(self):
        
        """
        Initializes a new inventory object with an empty list of products.
        """
        
        self.products = []
        self.guadagno = 0
        self.guadagno_netto = 0



    def chgIfInside(self,name,qt):
        for prod in self.products:
            if name==prod.name:
                prod.quantity=int(prod.quantity+qt)
                return True
            

    def register_product(self, product):
        
        """
        Register a new product in inventory.
        
        Args:
            product(Product): product to register.
        """
        
        self.products.append(product)



        
    def list_products(self):
        
        """
        Shows all products in inventory.
        """
        if not self.products:
            print("Non ci sono prodotti nell'inventario.")
        else:
            for product in self.products:
                print(product)
    
    def show_price(self):
        
        """
        Shows the gross and the net prices of all products in inventory.
        """
        if not self.products:
            print("Non ci sono prodotti nell'inventario.")
        else:
            for product in self.products:
                print(f"{product.name}: Prezzo lordo - €{product.gross_prices()}, Prezzo netto - €{product.net_prices()}")
                
    
    def isInside(self,nome):
        for prod in self.products:
            if nome==prod.name:
                return True
        return False

    def vendita(self,name,qt):
        for prod in self.products:
            if name==prod.name:
                if int(qt)>int(prod.quantity):
                    print("La quantità inserita è maggiore di quella disponibile")
                    return 0
                guadagno=int(qt)*float(prod.sale_price)
                guadagno_netto = int(qt)*(float(prod.sale_price)-float(prod.purchase_price))
                prod.quantity = int(prod.quantity)-int(qt)
                self.guadagno = self.guadagno+guadagno
                self.guadagno_netto = self.guadagno_netto+guadagno_netto
                print(f"Hai quadagnato €{guadagno} di cui €{guadagno_netto} netti")
        
        


    def input_numero(self,str):
        while True:
            numero = input(str)
            if numero.replace('.', '', 1).isdigit() and numero.count('.') <= 1 and not numero.startswith('-'):
                return float(numero)
            else:
                print("Input non valido. Devi inserire un numero valido (positivo o zero).")


    def menu_help(self):
        
        """
        Display a help menu with all available commands and allows the user to execute the various methods of the class.
        """
        
        print("Help Menu:")
        print("1. Aggiungi.")
        print("2. Elenca.")
        print("3. Mostra i prezzi totali lordi e netti dei prodotti")
        print("4. Vendita")
        print("5. Guadagni.")
        print("6. Aiuto.")
        print("7. Chiudi.")
        
        while True:
            choice = input("\nInserisci il numero del comando desiderato: ").strip()
            
            if choice == "1":
                name = input("Nome del prodotto:")

                quantity = self.input_numero("Quantità: ")

                if self.chgIfInside(name,quantity):
                    print(f"Il prodotto {name} è già presente nel magazzino: Non è necassario riscrivere i prezzi")
                    print("La quantità è stata aggiornata")
                else:
                    purchase_price = self.input_numero("Prezzo di acquisto: ")
                
                    sale_price = self.input_numero("Prezzo di vendita: ")
                
                    new_product = Product(name, int(quantity), purchase_price, sale_price) 
                
                    self.register_product(new_product)
                    print(f"Il prodotto {name} è stato registrato nell'inventario.")
                
                
            elif choice == "2":
                print("Prodotti nell'inventario:")
                self.list_products()
                
            elif choice == "3":
                print("\nPrezzi totali lordi e netti:")
                self.show_price()

            elif choice == "4":       
                nome = input("Nome prodotto da vendere: ")
                if(self.isInside(nome)):
                    qt = input("Quantità da vendere: ")
                    self.vendita(nome,qt)

            elif choice == "5":
                print(f"Guadagno totale lordo: €{self.guadagno} - Guadagno totale netto: €{self.guadagno_netto}")

            elif choice == "6":
                print("Help Menu:")
                print("1. Aggiungi.")
                print("2. Elenca.")
                print("3. Mostra i prezzi totali lordi e netti dei prodotti")
                print("4. Vendita")
                print("5. Guadagni.")
                print("6. Aiuto.")
                print("7. Chiudi.")
                
            elif choice == "7":
                print('Salvataggio...')
                with open("magazzino.pkl", 'wb') as file:
                    pickle.dump(self,file)
                print("Exit.")

                break
                
            else:
                print("Opzione non valida")



def main():
    print("Creazione inventario...")

    try:
        with open('magazzino.pkl','rb') as file:
            inventory=pickle.load(file)
            
            if len(inventory.products)>0:
                print("Prodotti nell'inventario:\n")
                inventory.list_products()
                print("\nPrezzi totali lordi e netti dei prodotti:\n")
                inventory.show_price()
    except:
        print("magazzino vuoto")
        inventory=Inventory()


    
    
    # Help Menu
    print("\n-------------------------------------------")
    inventory.menu_help()
    
if __name__ == "__main__":
    print('--- Execution of the function main ---')
    main()
```
